### Vector/Vector

* $\bf{x} \in {\rm I\!R^3}, \bf{y} \in {\rm I\!R^2}$, $f$ is represented in  graph (3):   
  
  * outputs $y_1$ and $y_2$ are both computed from inputs $x_1,x_2,x_3$;

  * partial derivatives between $y_1, y_2$ and $x_1,x_2,x_3$ are summarized by a matrix called **Jacobian**

    \begin{equation}
    \begin{bmatrix}
    \frac{\partial y}{\partial x}
    \end{bmatrix} =
    \begin{bmatrix}
    \frac{\partial y_1}{\partial x_1} & \frac{\partial y_1}{\partial x_2} & \frac{\partial y_1}{\partial x_3}\\
    \frac{\partial y_2}{\partial x_1} & \frac{\partial y_2}{\partial x_2} & \frac{\partial y_2}{\partial x_3}
    \end{bmatrix}.
    \end{equation}

![](https://drive.google.com/uc?export=view&id=1rTPIueuMyoQEUtr-kvGiLh3zmB0NrWAO)

### Chaining between Gradients and Jacobians 

* consider composite function $z = g(y)$ and $y = f(x)$ where $z \in {\rm I\!R}$:

   1. for $x, y \in {\rm I\!R}$, $g$ can be represented as 
   
      ![](https://drive.google.com/uc?export=view&id=1R5YxMMMKou69AHs66XG3iEy2F1cHUT9c)
      
     * Derivative of $z$ w.r.t $x$ is obtained with the chain rule 
     
     \begin{equation}
     \frac{dz}{dx} = \frac{dz}{dy} \times \frac{dy}{dx}.
     \end{equation}

* for vectorial $x,y$ for example $\bf{x} \in {\rm I\!R^3}, \bf{y} \in {\rm I\!R^2}$ the function $g$ would be represented as    
   
![](https://drive.google.com/uc?export=view&id=1b8EJlpwROp4jlx9UZNAEEfK6BP-BWdkS)
   
* Partial derivatives of $z$ w.r.t $x_1,x_2,x_3$ are obtained by the generalizations of chain rule
    
    \begin{equation}
    \frac{\partial z}{\partial x_1} = \left( \frac{\partial z}{\partial y_1} \times \frac{\partial y_1}{\partial x_1} \right) + \left( \frac{\partial z}{\partial y_2} \times \frac{\partial y_2}{\partial x_1} \right) = \sum_{j = 1}^2  \frac{\partial z}{\partial y_j} \times \frac{\partial y_j}{\partial x_1}
    \end{equation}
    





* In other words the above chain rule first computes  partial derivatives of $y_1, y_2$ w.r.t $x_1$ multiplied by the partial derivative of $z$ w.r.t $y_j$ and sum everything. Same principle applies for other partial derivatives of $z$ w.r.t $x_2,x_3$
     
     \begin{equation}
     \frac{\partial z}{\partial x_2} = \sum_{j = 1}^2  \frac{\partial z}{\partial y_j} \times \frac{\partial y_j}{\partial x_2},\\
     \frac{\partial z}{\partial x_3} = \sum_{j = 1}^2  \frac{\partial z}{\partial y_j} \times \frac{\partial y_j}{\partial x_3}.\\
     \end{equation}

## Loss Function (Mean Squared Error)

### Model

* Simple linear regression is defined as  

\begin{equation}        
y = w_1 x + b,
\end{equation}

* where $w_1$ and $b$:

  * respectively represent slope and  intercept of the line 

  * and whose values are **estimated** (or **learned**)  from  set of $n$ observations $\chi = \{(x^{(1)},y^{(1)}),\ldots, (x^{(n)},y^{(n)})\}$;


* Parameters are packaged into one row vector $\bf{\theta}$ $= [b, w_1]$ and estimated by optimizing some peformance criteria called **Loss Fonction**.



* Loss functions summarize model performance (according to its parameters $\bf{\theta}$) over  $\chi = \{(x^{(1)},y^{(1)}),\ldots, (x^{(n)},y^{(n)})\}$;



* Estimation of parameters is obtained by the **minimization** of $J(\bf{\theta})$;


* In regression like models, $J$ is defined as the mean of errors squared 


\begin{equation}
J(\bf{\theta}) = \frac{1}{n}\sum_{i = 1}^{n}({y}^{(i)} - \hat{y}^{(i)} )^2,
\end{equation}

* where $\hat{y}^{(i)} =  w_1 x^{(i)} + b$ is the output of $x^{(i)}$ obtained with  $\bf{\theta}$.

## Estimation of Parameters

### Analytical Approach:

* Linear regression can be extended to $d$ dimensional inputs

\begin{equation}
y^{(i)} = w_1 x_1^{(i)} +\cdots + w_d x_d^{(i)} + b.
\end{equation}


* Let $\bf{X}$ be $n \times \left(d+1\right)$ matrix including all data inputs and $\bf{y} \in {\rm I\!R}$ the associated outputs where


\begin{equation}
\bf{X} = 
\begin{bmatrix}
    1      & x_{1}^{(1)} & \dots  & x_{d}^{(1)} \\
    1      & x_{1}^{(2)} & \dots  & x_{d}^{(2)} \\
    \vdots & \vdots & \ddots & \vdots \\
    1      & x_{1}^{(n)} & \dots  & x_{d}^{(n)}
\end{bmatrix}
\end{equation}


* This linear model can be written in the following compact form


\begin{equation}
\bf{y} = \bf{X} \bf{\theta},
\end{equation}

where 
\begin{equation}
\bf{\theta} =
\begin{bmatrix}
b\\
w_1\\
\vdots\\
w_d
\end{bmatrix} 
\end{equation}


* Vector of parameters $\bf{\theta}$ can be obtained analytically as follows


\begin{equation}
\bf{\hat{\theta}} = (\bf{X^{T}X})^{-1}\bf{X^T}\bf{y}.
\end{equation}


* Because of computation cost relevant to matrix inversions, analytical approach become impractical in big datasets;


* **Gradient-Based** approaches become the default alternative to iteratively minimize $J(\bf{\theta})$;



#### Quick Recall on Gradient Descent



* Gradient Descent is fed with data $\bf{X}, \bf{y}$ and iteratively minimizes $J$ by updating each parameter $w_j$  

\begin{equation}
w_{j}^{[i]} \leftarrow w_{j}^{[i-1]} - \delta \frac{\partial J}{\partial w_{j}^{[i-1]}}. 
\end{equation}


* where $\delta > 0$ as the **learning rate** that defines the step by which the value of the parameter is altered.

\begin{equation}
J(\bf{w}) = \frac{1}{n}\sum_{i = 1}^{n}(\hat{y}^{(i)} - y^{(i)} )^2 = \frac{1}{n}\left[(w_1 x_1^{(1)} + b - y^{(1)} )^2 + \cdots + (w_1 x_1^{(n)} + b - y^{(n)} )^2 \right].

```
import numpy as np
import decimal
import matplotlib.pyplot as plt

def gradient_J(X,y,w):
    return(np.array([
            (2/len(X))*np.sum((w[0]) + (w[1]*X[:,1]) - y),
            (2/len(X))*np.sum((np.power(X[:,1],2)*w[1]) - (y*X[:,1]) + (w[0]*X[:,1]))
                   ]))
```

#### 3. Python Implementation of Loss Function (MSE)

```
def mse(w, X, Y):
    return (1/len(X)) * np.sum(np.square(Y - (w[0] + w[1]*X)))
```

#### 4. Python Implementation of  Gradient Descent Algorithm

```
decimal.getcontext().rounding = decimal.ROUND_DOWN   

x = np.array(meal_df['number_persons'].tolist())
y = np.array(meal_df['number_ordered_meals'].tolist())


X = np.column_stack([np.ones(len(x)), x])
theta = np.array([-10,10]) 

N = 2000
delta = 0.01

mse_track = np.zeros(N)
w_track = np.zeros([N,2])


for i in range(0, N):
    theta = theta - delta * gradient_J(X,y,theta)
    
    mse_track[i] = mse(theta, X[:,1], y)
    w_track[i] = theta

mse_track = np.array([float(round(decimal.Decimal(mse),3)) for mse in mse_track])
```

```
plt.figure(figsize = (20,8))
plt.subplot(1,2,1)
plt.rcParams['axes.facecolor'] = 'orange'
plt.grid(c='white')
idx_min = np.argmin(mse_track)
plt.plot(np.arange(0,N), mse_track, 'blue')
plt.xlabel('Iteration')
plt.ylabel('MSE')
plt.title('Minimum MSE = {:.4f} reached at iteration {}'.format(min(mse_track),idx_min+1), fontsize = 15)
plt.axvline(x = idx_min, color = 'green')

  
plt.subplot(1,2,2)
plt.rcParams['axes.facecolor'] = 'lightsteelblue'
plt.grid(c='white')
line_x = np.arange(0,max(x)+2)
w_hat = w_track[idx_min]
plt.plot(x, y, 'ro', line_x, w_hat[1]*line_x + w_hat[0], 'g')
plt.xlabel('Number of Clients')
plt.ylabel('Number of Ordered Meals')
plt.title('Estimated by Gradient-Descent', fontsize = 15)
plt.show()
```

### **Regression Metrics**

#### **Root Mean Squared Error (RMSE)** 

* RMSE measures the average deviation w.r.t real target in its original unit.

\\

\begin{equation}
RMSE(\bf{\theta}) = \sqrt(\frac{1}{n}\sum_{i = 1}^{n}({y}^{(i)} - \hat{y}^{(i)} )^2).
\end{equation}

#### **Coefficient of Determination (r$^2$)**

*  goodness of fit criterion defined into $]- \infty, 1]$;

* The closer to $1$ the better the goodness of fit is;

\\

\begin{equation}
\bf{r}^2(\theta) = 1 - \frac{\mbox{SS}_{res}}{\mbox{SS}_{tot}},
\end{equation}

\\

* where

$$
\mbox{SS}_{res} = \sum_{i = 1}^{n} (y_i - \hat{y}_i)^2, \quad \mbox{SS}_{tot} = \sum_{i = 1}^{n} (y_i - \bar{y})^2. 
$$

# Polynomial Regression

## Quadratic Regression

* Let $x, y \in {\rm I\!R}$, linear regression is defined as

\begin{equation}        
y = w_1 x + b.
\end{equation}

 

* Quadratic regression is defined as

\begin{equation}        
y = w_1 x + w_2 x^2 + b,
\end{equation}

* Note that the above model models **quadratic** function of $x$ but remains **linear** function of the **parameters**. 

## Polynomial Regressions

* Powers of $x$ can be added to implement cubic and other **higher degrees** polynomials;


* Higher degree regression may represent an infinite set of functions as a result of Taylor's Series;


* When inputs are vectorial, polynomial regression includes **all interaction terms**;



### Example

* Let $\bf{x}$ $= [x_1, x_2]^T$ and $y$ be a scalar;


* Quadratic regression between $y$ and $\bf{x}$ is defined as

\begin{equation}        
y = w_1 x_1 + w_2 x_2 + w_3 x_1^2 + w_4 x_2^2  + w_5 x_1 x_2 + w_6 x_1^2 x_2 + w_7 x_1 x_2^2 + w_8 x_1^2 x_2^2 + b,
\end{equation}

* which still remains **linear** function of the parameters.

### Capacity, Underfitting and Overfitting 


* The purpose of a trained model is to perform **well** on **previously unseen inputs**;


* The process of learning consists in optimizing the **training error**;


* What separates ML from regular optimization problems is that we want **also** to optimize the **test error** (or **generalization error**); 


![](https://drive.google.com/uc?export=view&id=1wmlgE77_zKN-YN4JJVL0G0dEjswh0mG-)




### Capacity

* Model's capacity consists in its ability to fit a wide variety of functions;


* Models with low capacity may struggle to fit the training set;


* Models with high capacity tend to model random deviations by memorization instead of capturing deterministic relationships;


* Models with high capacity result in poor generalization;


#### Underfitting:
   * when a model cannot fit training set resulting into high **training error**;
   
   
   * this type of models are said to be **biased**.




![](https://drive.google.com/uc?export=view&id=1wmlgE77_zKN-YN4JJVL0G0dEjswh0mG-)

#### Overfitting:
   * when models reach low **training error** but high **generalization error**;
   
   
   * this type of models are said to have **high variance**.




![](https://drive.google.com/uc?export=view&id=1wmlgE77_zKN-YN4JJVL0G0dEjswh0mG-)

#### Balanced Model:
   * Balanced models aim to lower both training and generalization errors;
   

   * Underfitting usually comes from an over simplification of the relationships between inputs and outputs;


   * Conversly, Overfitting results from complex models.  




## Regularization

* Regularization is any modification made on
the learning algorithm to reduce its generalization error;


* Regulization may reduce generalization error while increasing training error;


* Popular way is to modify the loss function with a **penality term**;


* In a regression setting, **regulizared loss function** has the form   


\begin{equation}
J(\bf{w}) = \mbox{MSE}_{train} + \lambda R(\bf{W}),
\end{equation}

* where $R(\bf{W})$ is a function of the weights (without bias) and $\lambda$ the coefficient associated to the regularization. 

### $L_2$ Regularization


* Known as Ridge Regularization or simply as $L_2$ norm is defined as 


\begin{equation}
R(\bf{W}) = \bf{W}^T \bf{W}.
\end{equation}


* When added to $\mbox{MSE}_{train}$, the minimum of regularized loss $J(\bf{w})$ remains **unique**;


* It penalizes dimensions with high contributions and distributes the prediction over all elements of $\bf{w}$;


* $L_2$ performs well when output variable depends on **all** input variables; 


* However $L_2$ is not robust to outliers.



### $L_1$ Regularization


* Known as LASSO or $L_1$ norm is defined as 


\begin{equation}
R(\bf{W}) = \sum_{k = 1}^{d} \vert w_k \vert.
\end{equation}


* When added to $\mbox{MSE}_{train}$, the minimum of regularized loss $J(\bf{w})$  **is not unique**;


* $L_1$ produces sparse solution, that is most of the elements of $\bf{w}$ are equal to zero;


* Sparsity may be used as feature selection technique;


* $L_1$ tend to be robust to outliers; 


* $L_1$ produce models simple and interpretable but cannot learn complex patterns;

* Fit linear, quadratic and cubic regressions on the data;



* Which one is the more adequate ?

### 1. Fit and Visualize Linear Regression

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X,y = np.array(dataset_df['x']).reshape(-1,1), np.array(dataset_df['y']).reshape(-1,1)

lm = LinearRegression()
lm.fit(X,y)

y_hat = lm.predict(X)

mse = mean_squared_error(y,y_hat)

plt.rcParams['axes.facecolor'] = 'lightsteelblue'
plt.grid(c='white')
plt.plot(X, y_hat, color='blue')
plt.scatter(X, y, color='red')
plt.title("Linear Model. MSE = " + '%.2f'%(mse))
```

### 2: Fit and Visualize Quadratic Regression

```
from sklearn.preprocessing import PolynomialFeatures 

# Generate new features as x**0 and x**2 
poly_reg_2 = PolynomialFeatures(degree = 2)
X_2 = poly_reg_2.fit_transform(X)

#Fit parabolic regression on new X (X_2)
lm_2 = LinearRegression()
lm_2.fit(X_2, y)

#Predict
y_hat_2 = lm_2.predict(X_2)
mse2 = mean_squared_error(y,y_hat_2)

dataset_df_2 = pd.DataFrame({'x': np.concatenate([np.array(i) for i in X]),
                             'y_hat_2':  np.concatenate([np.array(i) for i in y_hat_2])})
dataset_df_2 = dataset_df_2.sort_values('x')

#Visualize
plt.rcParams['axes.facecolor'] = 'lightsteelblue'
plt.grid(c='white')
plt.plot(dataset_df_2['x'], dataset_df_2['y_hat_2'] , color='green')
plt.scatter(X, y, color='red')
plt.title("Quadratic Model. MSE = " + '%.2f'%(mse2))
print(mse/mse2)
```

### 3: Fit and Visualize Cubic Regression

```
# Generate new features as x**0, x**2 and x**3
poly_reg_3 = PolynomialFeatures(degree = 3)
X_3 = poly_reg_3.fit_transform(X)

#Fit parabolic regression on new X (X_3)
lm_3 = LinearRegression()
lm_3.fit(X_3, y)

#Predict
y_hat_3 = lm_3.predict(X_3)
mse3 = mean_squared_error(y,y_hat_3)

dataset_df_3 = pd.DataFrame({'x': np.concatenate([np.array(i) for i in X]),
                             'y_hat_3':  np.concatenate([np.array(i) for i in y_hat_3])})
dataset_df_3 = dataset_df_3.sort_values('x')

#Visualize
plt.rcParams['axes.facecolor'] = 'lightsteelblue'
plt.grid(c='white')
plt.plot(dataset_df_3['x'], dataset_df_3['y_hat_3'] , color='green')
plt.scatter(X, y, color='red')
plt.title("Quadratic Model. MSE = " + '%.2f'%(mse3))
print(mse/mse3)
```
